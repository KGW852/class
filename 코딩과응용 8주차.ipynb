```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

path = r'C:\Users\KGW\OneDrive\수업자료(학부)\코딩과응용\Python\iris data'
```

```python
# iris 데이터를 판다스로 불러옵니다.
iris = pd.read_csv(os.path.join(path, 'iris_data.csv'))
iris.head()
```

```python
# 'setosa', 'versicolor'클래스의 1~20행 'petal' 값을 추출합니다.
setosa_length = iris.loc[iris['class'] == 'Iris-setosa', 'petal_length'].iloc[:50].to_list()
setosa_width = iris.loc[iris['class'] == 'Iris-setosa', 'petal_width'].iloc[:50].to_list()
versicolor_length = iris.loc[iris['class'] == 'Iris-versicolor', 'petal_length'].iloc[:50].to_list()
versicolor_width = iris.loc[iris['class'] == 'Iris-versicolor', 'petal_width'].iloc[:50].to_list()

# 추출된 값 확인
print(setosa_length)
print(setosa_width)
print(versicolor_length)
print(versicolor_width)
```

```python
# 'setosa', 'versicolor'의 x축 length, y축 width로 산점도 그리기.
plt.scatter(setosa_length, setosa_width)
plt.scatter(versicolor_length, versicolor_width)
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```

```python
# 'setosa', 'versicolor'의 리스트 합치기
length = setosa_length + versicolor_length
width = setosa_width + versicolor_width
print(length)
print(width)
```

```python
# 'length', 'width'의 리스트 재생성
iris_data = [[l, w] for l, w in zip(length, width)]
print(iris_data)
```

```python
# 정답 데이터 세트 생성.
iris_target = [1] * 50 + [0] * 50
print(iris_target)
```

```python
# K-최근접 이웃
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(iris_data, iris_target)
```

```python
kn.score(iris_data, iris_target)
```

```python
# 새로운 데이터 예측.
plt.scatter(setosa_length, setosa_width)
plt.scatter(versicolor_length, versicolor_width)
plt.scatter(3.8, 1.5, marker='^')
plt.grid()
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```

```python
kn.predict([[3.8, 1.5]])
```

```python
# 넘파이 2차원 배열형태로 변환
input_arr = np.array(iris_data)
target_arr = np.array(iris_target)
print(input_arr)
```

```python
print(input_arr.shape)
```

```python
# 훈련세트와 테스트세트 분할을 위한 인덱스 생성
np.random.seed(42)
index = np.arange(99)
np.random.shuffle(index)

print(index)
```

```python
# 인덱스 적용 전 0번째 배열
print(input_arr[[0]])
```

```python
# 훈련세트 (값 + 정답) 만들기
train_input = input_arr[index[:70]]
train_target = target_arr[index[:70]]

print(train_input[9])
```

```python
# 테스트세트 (값 + 정답) 만들기
test_input = input_arr[index[70:]]
test_target = target_arr[index[70:]]
```

```python
# 시각화
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(test_input[:,0], test_input[:,1])
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```

```python
# 새로운 kn 모델 생성
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
# predict로 test세트 확인
kn.predict(test_input)
```

```python
print(test_target)
```

```python
# 새로운 데이터 예측.
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(2.4, 1.3, marker='*')
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```

```python
# 새로운 데이터 예측
kn.predict([[2.4, 1.3]])
```

```python
# 5개의 이웃 찾기
distances, indexes = kn.kneighbors([[2.4, 1.3]])
```

```python
# 새로운 데이터 이웃 확인하기.
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(2.4, 1.3, marker='*')
plt.scatter(train_input[indexes, 0], train_input[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```

```python
# 스케일 동일하게 맞추기
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(2.4, 1.3, marker='*')
plt.scatter(train_input[indexes, 0], train_input[indexes, 1], marker='D')
plt.xlim((0, 5))
plt.ylim((0, 5))
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```

```python
# 평균 및 표준편차 구하기
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
```

```python
# 표준점수 구하기
train_scaled = (train_input - mean) / std
print(train_scaled.shape)
```

```python
# 샘플의 표준점수 구하기
sample = ([2.4, 1.3] - mean) / std

# 표준점수로 변환한 훈련세트 산점도 작성
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(sample[0], sample[1], marker='*')
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```

```python
# k최근접이웃을 사용한 새로운 모델 훈련
kn.fit(train_scaled, train_target)
```

```python
# 테스트세트 표준점수 변환
test_scaled = (test_input - mean) / std

# 테스트세트에서 모델 평가
kn.score(test_scaled, test_target)
```

```python
print(kn.predict([sample]))
```

```python
# 5개의 이웃 찾기
distances, indexes = kn.kneighbors([sample])

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(sample[0], sample[1], marker='*')
plt.scatter(train_scaled[indexes, 0], train_scaled[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('width')
plt.title('setosa(1) versicolor(0) scatter')
plt.show()
```
